```python
import torch
import torch.nn as nn

import math
import torch.autograd as autograd

import re

import matplotlib.pyplot as plt
```

```python
# read string from text file named result.txt
def read_result():
    with open('result_3l_quasi.txt', 'r') as file:
        data = file.read()
    return data

# extract train and validation loss from string 

def extract_losses(log_string):
    train_losses = []
    val_losses = []
    
    patterns = {
        'train_loss': r'Epoch \d+/\d+, Train Loss: ([\d\.]+)',
        'val_loss': r'Epoch \d+/\d+, Val Loss: ([\d\.]+)'
    }
    
    for line in log_string.split('\n'):
        print(line)
        train_loss_match = re.search(patterns['train_loss'], line)
        val_loss_match = re.search(patterns['val_loss'], line)
        
        if train_loss_match:
            print("train loss matched")
            train_losses.append(float(train_loss_match.group(1)))
        
        if val_loss_match:
            print("val loss matched")
            val_losses.append(float(val_loss_match.group(1)))
    
    return train_losses, val_losses
```

```python
train_losses, val_losses = extract_losses(read_result())

# draw the loss curve using matplotlib 
plt.plot(train_losses, label='Train Loss')
plt.plot(val_losses, label='Validation Loss')
plt.legend()

# zero margins
plt.margins(x=0, y=0)

# make the background transparent
plt.gcf().patch.set_facecolor('white')
plt.gcf().patch.set_alpha(0.0)



# plt.show()

# print the final train and validation loss
print("Final Train Loss: ", train_losses[-1])
print("Final Validation Loss: ", val_losses[-1])

# save the plot to a file 
plt.savefig('loss_curve_3l_quasi.png')


```
